```python
import cv2
import glob
import os
cam = cv2.VideoCapture(0)

face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

count = 0
while(True):
    ret, img = cam.read()
    #gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(img, 1.3, 5)
    for (x,y,w,h) in faces:
        x1 = x
        y1 = y
        x2 = x+w
        y2 = y+h
        test = img[y1:y2,x1:x2]
        cv2.rectangle(img, (x1,y1), (x2,y2), (255,255,255), 2)     
        count += 1
        # Save the captured image into the datasets folder
        
        test = img[y1:y2,x1:x2]
        print(type(test))
        print(test.shape)
        cv2.imwrite("dany_only/dany" + str(count) + ".jpg", test)
        
        cv2.imshow('image', img)
            
        
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    elif count >= 20: # Take 30 face sample and stop video
         break
cam.release()
cv2.destroyAllWindows()
```

```python
# jetson for
import jetson.inference
import jetson.utils
import cv2
import glob
import os
net = jetson.inference.imageNet('',['--model=models/my_data_50/resnet18.onnx','--labels=data/my_data/labels.txt','--input_blob=input_0','--output_blob=output_0'])
cam = cv2.VideoCapture(0)

face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

count = 0
while(True):
    ret, img = cam.read()
    #gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(img, 1.3, 5)
    for (x,y,w,h) in faces:
        x1 = x
        y1 = y
        x2 = x+w
        y2 = y+h
        test = img[y1:y2,x1:x2]
        test = jetson.utils.cudaFromNumpy(test)
        class_idx, confidence = net.Classify(test)
        class_desc = net.GetClassDesc(class_idx)
        print('JIO MERE',class_desc)
        cv2.rectangle(img, (x1,y1), (x2,y2), (255,255,255), 2)
        #cv2.putText(img, str(str(round(confidence*100)) + class_desc), (x+5,y-5), font, 1, (255,0,0), 4)
        count += 1
        
        cv2.imshow('image', img)
            
        
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    elif count >= 100: # Take 100 face sample and stop video
         break
cam.release()
cv2.destroyAllWindows()
```
